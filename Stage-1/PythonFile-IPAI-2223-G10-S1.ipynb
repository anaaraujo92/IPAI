### Mudança de nome das variáveis

```
def rename_years(df):
    columns = df.columns
    new_columns = []
    for column in columns:
        try:
            int(column.split(" ")[0])
            if len(column.split(" ")[0]) == 4:
                x = column.split(" ")
                new_columns.append(x[0])
        except ValueError:
            new_columns.append(column)
        
    for i, column in enumerate(columns):
        df.rename(columns = {column:new_columns[i]}, inplace = True)
```

### Adicionar Colunas

```
#Adicionar os countrys_codes de cada respetivo país ao dataset agricultura
country_df = df_gdp[['Country_code', 'Country']]
country_df2 = df_desastres[['Country_code', 'Country']]
result = pd.concat([country_df, country_df2], ignore_index=True)
result.drop_duplicates(inplace=True)
country_desastres = list(result.itertuples(index=False, name=None))
```

```
for i in range(0, len(country_desastres)):
    df_agricultura.loc[df_agricultura['Country'] == country_desastres[i][1], 'Country_code'] = country_desastres[i][0]
```

```
#VERIFICAR COUNTRY_CODES OKAY
mask = ~df_agricultura['Country_code'].apply(lambda x: isinstance(x, int))

int_values = df_agricultura.loc[~mask, 'Country_code'].unique()
print(int_values)
```

```
#Decidimos atribuir a China, mainland que não existia em mais nenhum dataset o country_code de CHM
df_agricultura.loc[df_agricultura['Country'] == 'China, mainland', 'Country_code'] = 'CHM'
```
